#### Import necessary libraries

```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from NN_models import *
%matplotlib inline
```

#### Import dataset and data analysis and pre-processing can be found at https://github.com/harikachandamollu/heart-disease-NN-one-hidden-layer. 

```python
data = np.genfromtxt('heart.csv',delimiter=',')

X=np.array((data[1:,:-1]))
Y=np.array((data[1:,-1]))
Y = np.reshape(Y,(Y.shape[0],1))

train_X, test_X, train_Y, test_Y = train_test_split(X,Y,test_size=0.2,random_state=0)
train_X=train_X.T
test_X=test_X.T
train_Y=train_Y.T
test_Y=test_Y.T

print("Shape of X(train_set):",train_X.shape)
print("Shape of X(test_set):",test_X.shape)
print("Shape of Y(train_set):",train_Y.shape)
print("Shape of Y(test_set):",test_Y.shape)

print('number of training examples:',X.shape[0])
print('number of input features:',X.shape[1])
print('number of target variables:',Y.shape[1])
```

#### From the above information, number of units in the input layer is 13 and number of units in the output layer is 1. For now, lets select random number of hidden units in hidden layer. 

```python
act_list=["relu","tanh"]
layer_dims_list=[[13,1],
                [13,5,1],
                [13,5,3,1],
                [13,5,3,2,1]]
learning_rate_list=[0.001,0.025,0.05,0.075,0.1]
```

### Using above hyperparameters, plotting the models to decide which combination gives better accuracy

```python
train_accuracy=[]
test_accuracy=[]
function="relu"  
for layer_dims in layer_dims_list:
    print(layer_dims)
    parameters = L_layer_model(train_X,train_Y,layer_dims,num_iterations=2500,print_cost=True,act=function)
    pred_train, accuracy = predict(train_X,train_Y,parameters,act=function)
    train_accuracy.append(accuracy)
    pred_test, accuracy = predict(test_X,test_Y,parameters,act=function)
    test_accuracy.append(accuracy)
    print("-----------------------------------------------------------------------------------------------")
plt.plot(train_accuracy,'-b',label="train_accuracy")
plt.plot(test_accuracy, '-r',label="test_accuracy")
plt.xlabel('number of hidden layers')
plt.ylabel('accuracy')
plt.title('learning rate:0.075(default) and activation_function:relu')
plt.legend()
plt.show()
```

```python
train_accuracy=[]
test_accuracy=[]
function="tanh"  
for layer_dims in layer_dims_list:
    print(layer_dims)
    parameters = L_layer_model(train_X,train_Y,layer_dims,num_iterations=2500,print_cost=True,act=function)
    pred_train, accuracy = predict(train_X,train_Y,parameters,act=function)
    train_accuracy.append(accuracy)
    pred_test, accuracy = predict(test_X,test_Y,parameters,act=function)
    test_accuracy.append(accuracy)
    print("-----------------------------------------------------------------------------------------------")
plt.plot(train_accuracy,'-b',label="train_accuracy")
plt.plot(test_accuracy, '-r', label="test_accuracy")
plt.xlabel('number of hidden layers')
plt.ylabel('accuracy')
plt.title('learning rate: 0.075(default) and activation_function:tanh')
plt.legend()
plt.show()
```

__From the above graphs following conclusions are made(only subjected to this dataset and these hyperparameters).__

__1. One or two hidden layers gives better results for relu activation function.__

__2. Only one hidden layer gives better results for tanh activation function.__

__3. When opted for 2 hidden layers it is better to choose relu activation function.__

```python
function="relu"
for lr in learning_rate_list:
    print("learning rate:",lr)
    train_accuracy=[]
    test_accuracy=[]
    for layer_dims in layer_dims_list:
        print(layer_dims)
        parameters = L_layer_model(train_X,train_Y,layer_dims,learning_rate=lr,num_iterations=2500,print_cost=True,act=function)
        pred_train, accuracy = predict(train_X,train_Y,parameters,act=function)
        train_accuracy.append(accuracy)
        pred_test, accuracy = predict(test_X,test_Y,parameters,act=function)
        test_accuracy.append(accuracy)
        print("-----------------------------------------------------------------------------------------------")
    plt.plot(train_accuracy,'-b',label="train_accuracy")
    plt.plot(test_accuracy, '-r', label="test_accuracy")
    plt.xlabel('number of hidden layers')
    plt.ylabel('accuracy')
    plt.title('activation_function:relu')
    plt.legend()
    plt.show()
```

__From the above plots we can conclude that for activation function relu,__

__1. For learning rate, 0.01 - one hidden layer yields high accuracy.__

__2. For learning rates, 0.025 and 0.05 - two hidden layers yield high accuracy.__

__3. The highest accuracy is obtained with 0.025 learning rate with 2 hidden layers.__

```python
function="tanh"
for lr in learning_rate_list:
    print("learning rate:",lr)
    train_accuracy=[]
    test_accuracy=[]
    for layer_dims in layer_dims_list:
        print(layer_dims)
        parameters = L_layer_model(train_X,train_Y,layer_dims,learning_rate=lr,num_iterations=2500,print_cost=True,act=function)
        pred_train, accuracy = predict(train_X,train_Y,parameters,act=function)
        train_accuracy.append(accuracy)
        pred_test, accuracy = predict(test_X,test_Y,parameters,act=function)
        test_accuracy.append(accuracy)
        print("-----------------------------------------------------------------------------------------------")
    plt.plot(train_accuracy,'-b',label="train_accuracy")
    plt.plot(test_accuracy, '-r', label="test_accuracy")
    plt.xlabel('number of hidden layers')
    plt.ylabel('accuracy')
    plt.title('activation_function:tanh')
    plt.legend()
    plt.show()
```

__From the above plots, for tanh activation function, one and two hidden layers produce similar accuracy with learning rate 0..025,0.05 and 0.075. With learning rate 0.01, one hidden layer gives better accuracy. The highest accuracy is obtained with 2 hidden layers and 0.075 learning rate.__
